## Data preprocessing

```python
#converting test dataset  into lowercase words 
test_df['article']=test_df['article'].str.lower()
test_df['highlights']=test_df['highlights'].str.lower()
```

```python
#converting train dataset  into lowercase words 
train_df['article']=train_df['article'].str.lower()
train_df['highlights']=train_df['highlights'].str.lower()
```

```python
#converting validation dataset  into lowercase words 
validation_df['article']=validation_df['article'].str.lower()
validation_df['highlights']=validation_df['highlights'].str.lower()
```

```python
#Removing unwanted tags if amy present
import re
train_df['highlights'] = train_df['highlights'].str.replace(r'<[^<>]*>', '', regex=True)
train_df['article'] = train_df['article'].str.replace(r'<[^<>]*>', '', regex=True)
#test_df
test_df['highlights']=test_df['highlights'].str.replace(r'<[^<>]*>', '', regex=True)
test_df['article']=test_df['article'].str.replace(r'<[^<>]*>', '', regex=True)
#validation_df
validation_df['highlights']=validation_df['highlights'].str.replace(r'<[^<>]*>', '', regex=True)
validation_df['article']=validation_df['article'].str.replace(r'<[^<>]*>', '', regex=True)
```

```python
#Removing punctuation 
import string
exclude=string.punctuation
exclude
```

```python
def remove_punc(sentence):
    return sentence.translate(str.maketrans('','',exclude))
  
```

```python
#train_df
train_df['highlights']=train_df['highlights'].apply(remove_punc)
train_df['article']=train_df['article'].apply(remove_punc)
#test_df
test_df['highlights']=test_df['highlights'].apply(remove_punc)
test_df['article']=test_df['article'].apply(remove_punc)
#validation_df                                              
validation_df['highlights']=validation_df['highlights'].apply(remove_punc)
validation_df['article']=validation_df['article'].apply(remove_punc)
```

```python
#Removing stop wprds
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text=[]
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x) 
```
